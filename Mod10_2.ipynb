```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
x_100 = np.random.normal(3.0, 1.0, size=100)
x_10000 = np.random.normal(3.0, 1.0, size=10000)
```

```python
sns.distplot(x_100)
sns.distplot(x_10000)
```

Średnia:

```python
s_100 = np.mean(x_100)
print(f"x_100= {s_100}")
s_10000 = np.mean(x_10000)
print(f"x_10000= {s_10000}")
```

Mediana:

```python
med_100 = np.median(x_100)
print(f"x_100= {med_100}")
med_10000 = np.median(x_10000)
print(f"x_10000= {med_10000}")
```

Odchylenie standardowe:

```python
sd_100 = np.std(x_100, ddof=1)
print(f'x_100= {sd_100}')
sd_10000 = np.std(x_10000, ddof=1)
print(f'x_10000= {sd_10000}')
```

Wariancja:

```python
v_100 = np.var(x_100, ddof=1)
print(f'x_100= {v_100}')
v_10000 = np.var(x_10000, ddof=1)
print(f'x_10000= {v_10000}')
```

Zgodnie z otrzymanymi wynikami powyższych staystyk opisowych, zbiór liczniejszy "x_10000" jest zbiorem znacznie mniej rozproszonym.
